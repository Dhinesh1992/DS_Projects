```python
# importing required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# all the values from 'Unnamed: 32' has null values and 'id' column has unique entries. Hence we can drop those columns
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
df.dtypes
```

```python
# Converting the object column into integer
df.diagnosis.unique()
```

```python
# We can convert 'M' -> 1 and 'B' ->0
df['diagnosis']=np.where(df['diagnosis']=='M',1,0)
df.head()
```

```python
for i in df.drop('diagnosis',axis=1).columns:
    sns.distplot(df[i],hist=False)
    plt.show()
```

#### From the above diagram it is seen that all the columns almost follow normal distribution

```python
#Splitting the i/p and o/p features
X = df.drop('diagnosis',axis=1)
y = df['diagnosis'].copy()
```

```python
plt.figure(figsize=(30,27))
sns.heatmap(X.corr(),annot=True)
plt.show()
```

###### from the above plot, it is understood that there are many features that are highly correlated with each other. Hence we can remove one of the features from the highly correlated features

### Feature Selection

```python
def corr_features(df):
    df_corr = df.corr()
    df_cols = df_corr.columns
    corr_cols =set()
    for i in range(len(df_cols)):
        for j in range(i):
            if abs(df_corr.iloc[i,j]) >=0.5 and df_cols[i] not in corr_cols and df_cols[i] != df_cols[j]:
                corr_cols.add(df_cols[i])
    return corr_cols  
        
```

```python
X1 = X.drop(list(corr_features(X)),axis=1).copy()
```

```python
X1.head()
```

### Importing the required models for classification

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
```

```python
# Splitting train and test data
X_train,X_test,y_train,y_test = train_test_split(X1,y,test_size=0.3,random_state=101)
```

```python
# Initializing the classification models
lr = LogisticRegression()
dt = DecisionTreeClassifier(ccp_alpha=0.0)
knn = KNeighborsClassifier(n_neighbors=5)
sv = SVC()
nb = MultinomialNB()
```

```python
models = [lr,dt,knn,sv,nb]
scores = []
for i in models:
    scr = cross_val_score(i,X_train,y_train,cv=5)
    scores.append(np.mean(scr))

plt.figure(figsize=(10,6))
plt.plot([str(i) for i in models],scores,'r-',marker='o')
plt.xticks(rotation=90)
plt.text(0.0,scores[0]+0.005,'LR')
plt.text(1,scores[1]+.005,'DT')
plt.text(2,scores[2]+.005,'KNN')
plt.text(3,scores[3]+.005,'SVC')
plt.text(4,scores[4]+.005,'MultinomialNB')
plt.show()
```

### From the above diagram it's clear that SVC shows highest training score

```python
sv.fit(X_train,y_train)
y_pred = sv.predict(X_test)

accuracy_score(y_test,y_pred)
```

```python
dt.fit(X_train,y_train)
y_pred1 = dt.predict(X_test)

accuracy_score(y_test,y_pred1)
```

```python
lr.fit(X_train,y_train)
y_pred2 = lr.predict(X_test)
accuracy_score(y_test,y_pred2)
```
