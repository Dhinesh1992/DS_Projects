## ---------------------------------Webscraping product details from flipkart---------------------------------

```python
# importing required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import lxml
pd.set_option('display.max_rows',None)
```

```python
# Creating a function to get the url and proceed
def main(url,line_num):
    
    # Creating and opening file in 'append' mode
    file = open('flipkart_laptops.csv','a')
    
    # Creating a useragent. hence webserver doesn't think this is spam request
    HEADERS = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:47.0) Gecko/20100101 Firefox/47.0'}
    
    # Using requests passing the request to the webserver
    webpage = requests.get(url,headers=HEADERS)
    
    # Creating a soup object and parsing the content of the response page
    soup = BeautifulSoup(webpage.content,'html.parser')
    
    # Retrieving the required information from parsed webpage
    # Scraping product_title details
    try:
        # Extracting the title using tag and class information  from flipkart page
        title = soup.find('span',attrs={'class':'B_NuCI'}).text.strip().replace(',','')
        
        # Writing the title information into the file
        file.write(f'{title},')
        #print(title)
        
    except AttributeError:
        
        # Incase of attribute not found error storing the default value into title
        title='NAN'
        
        # Writing the default value into the file
        #file.write(f'{title},')
        
    
    # Scraping product_price details
    try:
        
        price = soup.find('div',attrs={'class':'_30jeq3 _16Jk6d'}).string[1:].replace(',','')
        
        file.write(f'{price},')
        #print(price)
        
    except AttributeError:
        
        price = 'NaN'
        #file.write(f'{price},')
        
    # Scraping product_rating details    
    try:
        rating = soup.find('div',attrs={'class':'_3LWZlK'}).text.strip().replace(',','')
        file.write(f'{rating},')
        #print(rating)
        
    except AttributeError:
        rating='NaN'
        #file.write(f'{rating}')
        
    # Scraping discount details
    try:
        discount = soup.find('div',attrs={'class':'_3Ay6Sb _31Dcoz'}).string.strip().replace(',','')
        file.write(f'{discount}\n')
        #print(discount)
        
    except AttributeError:
        discount ='NaN'
        #file.write(f'{discount}\n')
    file.close()
    print(f' {line_num} Records loaded.')
    
```

```python
if __name__=='__main__':
    # opening url file to access the url links
    file = open('product_urls.txt','r')
    length = len(file.readlines())
    
    # Setting the line_number to 1 -> n
    n = length - (length-1)
    
    # since readlines() used already resetting the position to the start of the file
    file.seek(0)
    # iterating over the urls
    for link in file.readlines():
        
        main(link,n)
        n+=1
        
```
